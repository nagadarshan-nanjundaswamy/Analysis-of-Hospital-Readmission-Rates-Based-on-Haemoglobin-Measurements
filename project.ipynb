```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('X_train.txt', sep=' ', header=None)
display(df)
x_train = np.array(df.values)
print (x_train.shape)
print (x_train)

df = pd.read_csv('y_train.txt', sep=' ', header=None)
display(df)
y_train = df.values
print (y_train.shape)
print (y_train)

df = pd.read_csv('X_test.txt', sep=' ', header=None)
display(df)
x_test = df.values
print (x_test.shape)
print (x_test)

df = pd.read_csv('y_test.txt', sep=' ', header=None)
display(df)
y_test = df.values
print (y_test.shape)
print (y_test)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
print (x_train.shape)
print (x_train)
```

```python
print (x_test.shape)
print (x_test)
```

```python
y_train = y_train.squeeze()
y_test = y_test.squeeze()
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
knn = KNeighborsClassifier()
hyper_param = {'n_neighbors':[3, 4, 5, 6, 7]}
knn_gs = GridSearchCV(knn, hyper_param, scoring='accuracy', cv=5).fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
knn.n_neighbors = knn_gs.best_params_['n_neighbors']
knn_scores = cross_val_score(knn, x_test, y_test, cv=10)
```

```python
print (knn_gs.best_params_)
print (np.mean(knn_scores))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
lr = LogisticRegression(solver='lbfgs', max_iter=10000)
lr.fit(x_train, y_train)
lr_scores = cross_val_score(lr, x_test, y_test, cv=10, scoring='accuracy')
```

```python
print (np.mean(lr_scores))
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
nb = GaussianNB()
nb.fit(x_train, y_train)
nb_scores = cross_val_score(nb, x_test, y_test, cv=10, scoring='accuracy')
```

```python
print (np.mean(nb_scores))
print (nb.score(x_test, y_test))
```

### Decision Tree

```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt = dt.fit(x_train, y_train)
dt_scores = cross_val_score(dt,x_test,y_test,cv=10,scoring='accuracy')
```

```python
print (np.mean(dt_scores))
```

### SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
lsvm = SVC()
hyper_param = {'C':np.linspace(1e-5, 1e2, 10)}
lsvm_gs = GridSearchCV(lsvm, hyper_param, scoring='accuracy', cv=5).fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
lsvm.C = lsvm_gs.best_params_['C']
svm_scores = cross_val_score(lsvm, x_test, y_test, cv=10)
```

```python
print (lsvm.C)
print (np.mean(svm_scores))
```

### MLP

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(random_state=0, max_iter=10000).fit(x_train, y_train)
print (mlp.score(x_test, y_test))
```
